## **4. Análisis de Tipos de Datos**

```python
print("📋 Tipos de datos por columna:\n")
print(df.dtypes)
print("\n" + "="*50)

num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()

print(f"\n🔢 Variables numéricas ({len(num_cols)}): {num_cols}")
print(f"\n📝 Variables categóricas ({len(cat_cols)}): {cat_cols}")
```

## **5. Análisis de Valores Faltantes**

```python
missing_data = pd.DataFrame({
    'Columna': df.columns,
    'Valores_Faltantes': df.isnull().sum(),
    'Porcentaje': (df.isnull().sum() / len(df) * 100).round(2)
}).sort_values('Valores_Faltantes', ascending=False)

missing_data = missing_data[missing_data['Valores_Faltantes'] > 0]

if len(missing_data) > 0:
    print("⚠️ Valores faltantes detectados:\n")
    print(missing_data.to_string(index=False))
else:
    print("✅ No hay valores faltantes en el dataset")
```

```python
if len(missing_data) > 0:
    fig = px.bar(missing_data, 
                 x='Porcentaje', 
                 y='Columna',
                 orientation='h',
                 title='Porcentaje de Valores Faltantes por Columna',
                 labels={'Porcentaje': 'Porcentaje (%)', 'Columna': 'Columna'},
                 color='Porcentaje',
                 color_continuous_scale='Reds')
    fig.update_layout(height=500, showlegend=False)
    fig.show()
```

## **6. Variable Objetivo**

El dataset está **balanceado** (50.4% True / 49.6% False): no se requieren técnicas de balanceo de clases.

> **Análisis estadístico profundo en NB02:** la relación entre cada variable y el target se cuantifica en el **Notebook 02 · Análisis de Variables vs Target** con pruebas chi-square y Mann-Whitney. Ahí se documentan las decisiones de feature engineering (qué conservar, transformar o descartar).

```python
target = 'Transported'

print(f"🎯 Distribución de la variable objetivo '{target}':\n")
print(df[target].value_counts())
print("\nPorcentajes:")
print(df[target].value_counts(normalize=True).mul(100).round(2))
```

```python
fig = make_subplots(rows=1, cols=2, 
                    specs=[[{'type':'bar'}, {'type':'pie'}]],
                    subplot_titles=(f'Distribución de {target}', f'Proporción de {target}'))

value_counts = df[target].value_counts()

fig.add_trace(
    go.Bar(x=value_counts.index.astype(str), y=value_counts.values, 
           marker_color=['#3498db', '#e74c3c'],
           showlegend=False),
    row=1, col=1
)

fig.add_trace(
    go.Pie(labels=value_counts.index.astype(str), values=value_counts.values,
           marker_colors=['#3498db', '#e74c3c']),
    row=1, col=2
)

fig.update_layout(height=500, showlegend=True, title_text=f"Análisis de {target}")
fig.update_xaxes(title_text=target, row=1, col=1)
fig.update_yaxes(title_text="Frecuencia", row=1, col=1)
fig.show()
```

## **7. Análisis de Variables Numéricas**

### **7.1 Distribuciones**

```python
if len(num_cols) > 0:
    num_cols_to_plot = [col for col in num_cols if col != target]
    
    for col in num_cols_to_plot:
        fig = px.histogram(df, x=col, nbins=30,
                          title=f'Distribución de {col}',
                          labels={col: col, 'count': 'Frecuencia'},
                          color_discrete_sequence=['skyblue'])
        fig.update_layout(height=400, showlegend=False)
        fig.show()
```

### **7.2 Detección de Outliers (Boxplots)**

```python
if len(num_cols) > 0:
    num_cols_to_plot = [col for col in num_cols if col != target]
    
    for col in num_cols_to_plot:
        fig = px.box(df, y=col,
                     title=f'Boxplot de {col}',
                     labels={col: col},
                     color_discrete_sequence=['lightcoral'])
        fig.update_layout(height=400, showlegend=False)
        fig.show()
```

### **7.3 Matriz de Correlación**

```python
if len(num_cols) > 1:
    correlation_matrix = df[num_cols].corr()
    
    fig = px.imshow(correlation_matrix,
                    text_auto='.2f',
                    color_continuous_scale='RdBu_r',
                    aspect='auto',
                    title='Matriz de Correlación - Variables Numéricas',
                    labels=dict(color="Correlación"))
    fig.update_layout(height=500, width=600)
    fig.show()
```

## **8. Análisis de Variables Categóricas**

### **8.1 Cardinalidad**

```python
if len(cat_cols) > 0:
    cat_cols_to_analyze = [col for col in cat_cols if col != target]
    
    if len(cat_cols_to_analyze) > 0:
        cardinality = pd.DataFrame({
            'Columna': cat_cols_to_analyze,
            'Valores_Unicos': [df[col].nunique() for col in cat_cols_to_analyze]
        }).sort_values('Valores_Unicos', ascending=False)
        
        print("📊 Cardinalidad de variables categóricas:\n")
        print(cardinality.to_string(index=False))
```

### **8.2 Distribución de Categorías**

```python
if len(cat_cols) > 0:
    cat_cols_to_plot = [col for col in cat_cols if col != target and df[col].nunique() <= 20]
    
    for col in cat_cols_to_plot:
        value_counts = df[col].value_counts()
        fig = px.bar(x=value_counts.index.astype(str), y=value_counts.values,
                     title=f'Distribución de {col}',
                     labels={'x': col, 'y': 'Frecuencia'},
                     color_discrete_sequence=['steelblue'])
        fig.update_layout(height=400, showlegend=False)
        fig.show()
```

## **9. Relación entre Variables y Target**

### **9.1 Variables Numéricas vs Target**

```python
if len(num_cols) > 0:
    num_cols_to_plot = [col for col in num_cols if col != target]
    
    for col in num_cols_to_plot:
        fig = px.box(df, x=target, y=col,
                     title=f'{col} vs {target}',
                     labels={target: target, col: col},
                     color=target,
                     color_discrete_sequence=['#2ecc71', '#e74c3c'])
        fig.update_layout(height=400)
        fig.show()
```

### **9.2 Variables Categóricas vs Target**

```python
if len(cat_cols) > 0:
    cat_cols_to_plot = [col for col in cat_cols if col != target and df[col].nunique() <= 10]
    
    for col in cat_cols_to_plot:
        crosstab_data = pd.crosstab(df[col], df[target], normalize='index') * 100
        
        fig = go.Figure()
        for transported_val in crosstab_data.columns:
            fig.add_trace(go.Bar(
                name=str(transported_val),
                x=crosstab_data.index.astype(str),
                y=crosstab_data[transported_val],
                marker_color='#3498db' if transported_val else '#e74c3c'
            ))
        
        fig.update_layout(
            title=f'{col} vs {target} (Proporción %)',
            xaxis_title=col,
            yaxis_title='Proporción (%)',
            barmode='stack',
            height=400
        )
        fig.show()
```
