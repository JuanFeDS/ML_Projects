## **1. Librerías**

## **2. Cargar Datos**

```python
df = pd.read_csv('../../data/raw/train.csv')
df['Transported_int'] = df['Transported'].astype(int)

print(f'Dataset: {df.shape[0]:,} filas x {df.shape[1]} columnas')
print(f'Target balance: {df["Transported"].mean():.1%} True  |  {1-df["Transported"].mean():.1%} False')
```

## **3. Variables Categóricas**

Para cada variable categórica se calcula:
- Tasa de `Transported=True` por categoría
- **Chi-square test**: evalúa si la distribución de `Transported` es independiente de la categoría
  - p-valor < 0.05 → la variable tiene relación estadísticamente significativa con el target

```python
def analizar_categorica(df, col, target='Transported_int'):
    """Tasa de Transported por categoría + chi-square test."""
    resumen = (
        df.groupby(col)[target]
        .agg(['mean', 'count'])
        .rename(columns={'mean': 'tasa_transported', 'count': 'n'})
        .sort_values('tasa_transported', ascending=False)
        .reset_index()
    )
    contingency = pd.crosstab(df[col], df['Transported'])
    chi2, p, _, _ = stats.chi2_contingency(contingency)
    return resumen, chi2, p
```

### **3.1 HomePlanet**

```python
col = 'HomePlanet'
resumen, chi2, p = analizar_categorica(df.dropna(subset=[col]), col)

print(resumen.to_string(index=False))
print(f'\nChi-square: {chi2:.2f}  |  p-valor: {p:.2e}  |  {"SIGNIFICATIVA" if p < 0.05 else "no significativa"}')

fig = px.bar(resumen, x=col, y='tasa_transported', text='n',
             title=f'{col} — Tasa de Transported por categoría',
             labels={'tasa_transported': 'Tasa Transported'},
             color='tasa_transported', color_continuous_scale='RdYlGn')
fig.add_hline(y=df['Transported_int'].mean(), line_dash='dash',
              annotation_text='Media global')
fig.update_layout(height=400, showlegend=False)
fig.show()
```

### **3.5 Cabin — Deck y Side**

`Cabin` se descompone en `Deck` (cubierta) y `Side` (lado del barco) para el análisis.

```python
df_cabin = df.dropna(subset=['Cabin']).copy()
df_cabin['Deck'] = df_cabin['Cabin'].str.split('/').str[0]
df_cabin['Side'] = df_cabin['Cabin'].str.split('/').str[2]

for col in ['Deck', 'Side']:
    resumen, chi2, p = analizar_categorica(df_cabin, col)
    print(f'--- {col} ---')
    print(resumen.to_string(index=False))
    print(f'Chi-square: {chi2:.2f}  |  p-valor: {p:.2e}  |  {"SIGNIFICATIVA" if p < 0.05 else "no significativa"}')
    print()

fig = make_subplots(rows=1, cols=2, subplot_titles=['Deck', 'Side'])

deck_res = df_cabin.groupby('Deck')['Transported_int'].mean().reset_index().sort_values('Transported_int', ascending=False)
side_res  = df_cabin.groupby('Side')['Transported_int'].mean().reset_index()

fig.add_trace(go.Bar(x=deck_res['Deck'], y=deck_res['Transported_int'],
                     marker_color='#3498db', name='Deck'), row=1, col=1)
fig.add_trace(go.Bar(x=side_res['Side'], y=side_res['Transported_int'],
                     marker_color='#e74c3c', name='Side'), row=1, col=2)
fig.add_hline(y=df['Transported_int'].mean(), line_dash='dash')
fig.update_yaxes(title_text='Tasa Transported', row=1, col=1)
fig.update_layout(height=400, showlegend=False,
                  title='Cabin (Deck y Side) — Tasa de Transported')
fig.show()
```

## **4. Variables Numéricas**

Para cada variable numérica se calcula:
- Distribución por clase (`Transported=True` vs `False`)
- **Mann-Whitney U test**: compara las distribuciones sin asumir normalidad
  - p-valor < 0.05 → las distribuciones difieren significativamente entre clases
- Correlación de Pearson con el target

```python
def analizar_numerica(df, col, target='Transported'):
    """Mann-Whitney U test + correlación con el target."""
    df_ok = df.dropna(subset=[col])
    grupo_true  = df_ok[df_ok[target] == True][col]
    grupo_false = df_ok[df_ok[target] == False][col]
    stat, p = stats.mannwhitneyu(grupo_true, grupo_false, alternative='two-sided')
    corr = df_ok[col].corr(df_ok['Transported_int'])
    return grupo_true, grupo_false, stat, p, corr
```

### **4.1 Age**

```python
col = 'Age'
g_true, g_false, stat, p, corr = analizar_numerica(df, col)

print(f'Media Transported=True : {g_true.mean():.1f}')
print(f'Media Transported=False: {g_false.mean():.1f}')
print(f'Correlación con target : {corr:.3f}')
print(f'Mann-Whitney p-valor   : {p:.2e}  |  {"SIGNIFICATIVA" if p < 0.05 else "no significativa"}')

fig = go.Figure()
fig.add_trace(go.Histogram(x=g_true,  name='Transported=True',
                           opacity=0.7, marker_color='#2ecc71', nbinsx=40))
fig.add_trace(go.Histogram(x=g_false, name='Transported=False',
                           opacity=0.7, marker_color='#e74c3c', nbinsx=40))
fig.update_layout(barmode='overlay', title=f'{col} — Distribución por clase',
                  xaxis_title=col, yaxis_title='Frecuencia', height=400)
fig.show()
```

### **4.2 Variables de Gasto**

```python
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

print('Variable         | Media True | Media False | Correlación | p-valor MW   | Significativa')
print('-' * 90)
for col in spending_cols:
    g_true, g_false, stat, p, corr = analizar_numerica(df, col)
    sig = 'SI' if p < 0.05 else 'no'
    print(f'{col:16s} | {g_true.mean():10.1f} | {g_false.mean():11.1f} | {corr:11.3f} | {p:12.2e} | {sig}')
```

```python
fig = make_subplots(rows=2, cols=3,
                    subplot_titles=spending_cols + [''],
                    vertical_spacing=0.12)

positions = [(1,1),(1,2),(1,3),(2,1),(2,2)]

for (r, c), col in zip(positions, spending_cols):
    g_true  = df[df['Transported'] == True][col].dropna()
    g_false = df[df['Transported'] == False][col].dropna()
    fig.add_trace(go.Box(y=g_true,  name='True',  marker_color='#2ecc71',
                         showlegend=(r==1 and c==1)), row=r, col=c)
    fig.add_trace(go.Box(y=g_false, name='False', marker_color='#e74c3c',
                         showlegend=(r==1 and c==1)), row=r, col=c)
    fig.update_yaxes(title_text=col, row=r, col=c)

fig.update_layout(height=600, title='Variables de Gasto — Distribución por clase',
                  boxmode='group')
fig.show()
```

## **5. Features Derivadas Clave**

Análisis de variables que se crearán en el feature engineering para validar que tienen sentido antes de construirlas.

### **5.1 TotalSpending**

```python
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
df['TotalSpending'] = df[spending_cols].fillna(0).sum(axis=1)
df['TotalSpending_Log'] = np.log1p(df['TotalSpending'])

for col in ['TotalSpending', 'TotalSpending_Log']:
    g_true, g_false, stat, p, corr = analizar_numerica(df, col)
    print(f'{col}: correlación={corr:.3f}  |  p-valor={p:.2e}')

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=['TotalSpending (original)', 'TotalSpending (log)'])

for col, r, c, color_t, color_f in [
    ('TotalSpending', 1, 1, '#2ecc71', '#e74c3c'),
    ('TotalSpending_Log', 1, 2, '#2ecc71', '#e74c3c')
]:
    g_t = df[df['Transported']==True][col]
    g_f = df[df['Transported']==False][col]
    fig.add_trace(go.Histogram(x=g_t, name='True',  opacity=0.7,
                               marker_color=color_t, nbinsx=50,
                               showlegend=(col=='TotalSpending')), row=1, col=c)
    fig.add_trace(go.Histogram(x=g_f, name='False', opacity=0.7,
                               marker_color=color_f, nbinsx=50,
                               showlegend=(col=='TotalSpending')), row=1, col=c)

fig.update_layout(barmode='overlay', height=400,
                  title='TotalSpending — Distribución por clase')
fig.show()
```

### **5.2 GroupSize (desde PassengerId)**

```python
df['TravelGroup'] = df['PassengerId'].str.split('_').str[0]
df['GroupSize']   = df.groupby('TravelGroup')['TravelGroup'].transform('count')

resumen, chi2, p = analizar_categorica(df, 'GroupSize')
print(resumen.to_string(index=False))
print(f'\nChi-square: {chi2:.2f}  |  p-valor: {p:.2e}  |  {"SIGNIFICATIVA" if p < 0.05 else "no significativa"}')

fig = px.line(resumen.sort_values('GroupSize'), x='GroupSize', y='tasa_transported',
              markers=True, text='n',
              title='GroupSize — Tasa de Transported por tamaño de grupo',
              labels={'tasa_transported': 'Tasa Transported', 'GroupSize': 'Tamaño del grupo'})
fig.add_hline(y=df['Transported_int'].mean(), line_dash='dash',
              annotation_text='Media global')
fig.update_layout(height=400)
fig.show()
```
